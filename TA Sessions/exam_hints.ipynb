```python
import numpy as np
%matplotlib nbagg
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
def get_circle(
        radius = 1.,
        resolution = 8,
        orientation = 'z'):
    phi = np.linspace(0, 2*np.pi, num = resolution, endpoint = False)
    points = np.zeros((resolution, 3), dtype = np.float)
    p0 = radius*np.cos(phi)
    p1 = radius*np.sin(phi)
    lines = set([(i, (i+1) % resolution) for i in range(resolution)])
    if orientation == 'z':
        points[:, 0] = p0
        points[:, 1] = p1
    elif orientation == 'x':
        points[:, 1] = p0
        points[:, 2] = p1
    elif orientation == 'y':
        points[:, 2] = p0
        points[:, 0] = p1
    return points, lines

def get_sphere(
        radius = 1.,
        resolution = 8,
        center = np.zeros(3, dtype = np.float)):
    points = []
    lines = []
    z = np.linspace(-(radius - radius/resolution),
                     (radius - radius/resolution),
                     resolution)
    n = 0
    for zz in z:
        p, l = get_circle(radius = np.sqrt(radius**2 - zz**2),
                          resolution = resolution,
                          orientation = 'z')
        points.append(p + np.array([0., 0., zz]))
        lines += [[i[0]+n, i[1]+n] for i in l]
        n += resolution
    for i in range(resolution):
        for j in range(resolution-1):
            lines.append([i+j*resolution, i + (j+1)*resolution])
    return ((np.concatenate(points, axis = 0) +
             center),
            lines)

def plot_segments(a, points, connections):
    for c in connections:
        a.plot(points[list(c), 0], points[list(c), 1], points[list(c), 2], marker = '.')
    return None
```

```python
v, c = get_sphere(radius = 0.5)


fig = plt.figure()
a = fig.add_subplot(111, projection='3d')

n = 3
plot_segments(a, v, c)
t = np.linspace(0, 1, 2**10)
z = t - 0.5
rad = np.sqrt(0.25 - z**2)
x = np.cos(t*n*2*np.pi)*rad
y = np.sin(t*n*2*np.pi)*rad

a.plot(x, y, z, color = (0, 0, 0))
```

```python
# find intersections with y = 0

def intersect_curve(c, value):
    # c is a 1D numpy array, representing an R->R function, and we want to
    # solve the equation "c = value"
    dist_from_value = c - value
    dist_product = dist_from_value[1:] * dist_from_value[:dist_from_value.shape[0]-1]
    jstar = np.where(dist_product <= 0)[0]
    that = []
    for j in jstar:
        that.append((value - c[j]) / (c[j+1] - c[j]))
    return jstar, np.array(that)

jj, tt = intersect_curve(y, 0)
```

```python
def ycut_curve3D(x, y, z, yvalue):
    """
        Takes a parametrized curve in 3D as input, given as 3 1D numpy arrays.
        Returns 2 lists, of the approximate values of x and z for which y = yvalue.
        Uses linear interpolation.
    """
    dist_from_value = y - yvalue
    dist_product = dist_from_value[1:] * dist_from_value[:dist_from_value.shape[0]-1]
    jstar = np.where(dist_product <= 0)[0]
    that = []
    for j in jstar:
        that.append((yvalue - y[j]) / (y[j+1] - y[j]))
    xx = []
    zz = []
    for i in range(jstar.shape[0]):
        j = jstar[i]
        tval = that[i]
        xx.append(x[j]*(1-tval) + x[j+1]*tval)
        zz.append(z[j]*(1-tval) + z[j+1]*tval)
    return xx, zz

ecs0, zee0 = ycut_curve3D(x, y, z, 0)
ecs1, zee1 = ycut_curve3D(x, y, z, 0.2)
```

\begin{align}
    f_x &= B \cos y + C \sin z \\
    f_y &= C \cos z + A \sin x \\
    f_z &= A \cos x + B \sin y \\
\end{align}

\begin{equation}
    \frac{d}{dt} \mathbf{x} = \mathbf{f}
\end{equation}

```python
#A = -0.2
#B = 1.4
#C = 1.5
A = 1.0
B = 0.9
C = 1.1

def ABC_rhs(x):
    return np.array([B*np.cos(x[1]) + C*np.sin(x[2]),
                     C*np.cos(x[2]) + A*np.sin(x[0]),
                     A*np.cos(x[0]) + B*np.sin(x[1])])
```

```python
def Euler(rhs, dt = 0.5, N = 8, x0 = None):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    for t in range(N):
        x[t+1] = x[t] + dt*rhs(x[t])
    return x

def cRK(rhs, dt = 0.5, N = 8, x0 = None):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    for t in range(N):
        k1 = rhs(x[t])
        k2 = rhs(x[t] + 0.5*dt*k1)
        k3 = rhs(x[t] + 0.5*dt*k2)
        k4 = rhs(x[t] + dt*k3)
        x[t+1] = x[t] + dt*(k1 + 2*(k2 + k3) + k4)/6
    return x

def get_evdt(
        initial_condition,
        h0, N0, ndivisions,
        method,
        rhs):
    epsilon = []
    tstep = [h0]
    sol = []

    N = N0
    sol.append(method(rhs,
                      dt = tstep[-1],
                      N = N,
                      x0 = initial_condition))
    for n in range(1, ndivisions + 1):
        tstep.append(h0*2.**(-n))
        N = N*2
        sol.append(method(rhs,
                          dt = tstep[-1],
                          N = N,
                          x0 = initial_condition))
        epsilon.append((sol[-1][-1] - sol[-2][-1]) /
                       np.max(np.abs(sol[-1])))

    return (np.array(tstep[:len(tstep)-1]),
            np.abs(epsilon))
```

```python
ic = np.random.random((3, 64))*2*np.pi
dt, err = get_evdt(ic,
                   0.5, 4, 6, cRK,
                   ABC_rhs)
```

```python
def ycut_curve3D_modulo(x, y, z, yvalue):
    """
        Takes a parametrized curve in 3D as input, given as 3 1D numpy arrays.
        Returns 2 lists, of the approximate values of x and z for which y = yvalue.
        Uses linear interpolation.
    """
    dist_from_value = y - yvalue
    dist_product = (dist_from_value[1:] *
                    dist_from_value[:dist_from_value.shape[0]-1])
    jstar = np.where(np.logical_and(dist_product <= 0,
                                    np.abs(y[1:] - yvalue) <= 2*np.pi/10))[0]
    that = []
    for j in jstar:
        that.append((yvalue - y[j]) / (y[j+1] - y[j]))
    xx = []
    zz = []
    for i in range(jstar.shape[0]):
        j = jstar[i]
        tval = that[i]
        xx.append(x[j]*(1-tval) + x[j+1]*tval)
        zz.append(z[j]*(1-tval) + z[j+1]*tval)
    return xx, zz
```

```python
A = -0.5
B = 1.0
C = -1.0

def ABC_rhs(x):
    return np.array([B*np.cos(x[1]) + C*np.sin(x[2]),
                     C*np.cos(x[2]) + A*np.sin(x[0]),
                     A*np.cos(x[0]) + B*np.sin(x[1])])

ic = np.random.random((3, 64))*2*np.pi
ic[1] = np.pi


# integration times are long enough if, in the section plot below,
# curves can be distinguished by eye (it's acceptable if they appear as
# dashed or dotted lines).
sol = cRK(ABC_rhs, 2.**(-4), 2**15, x0 = ic)
sol_mod = (sol) % (2*np.pi)
```

```python
import sympy as sp

def ycut_curve3D_modulo_Lag4(x, y, z, yvalue, dt = 1.0):
    """
        Takes a parametrized curve in 3D as input, given as 3 1D numpy arrays.
        Returns 2 lists, of the approximate values of x and z for which y = yvalue.
        Uses linear interpolation.
    """
    dist_from_value = y - yvalue
    dist_product = (dist_from_value[1:] *
                    dist_from_value[:dist_from_value.shape[0]-1])
    jtemp = np.where(np.logical_and(dist_product <= 0,
                                    np.abs(y[1:] - yvalue) <= 2*np.pi/10))[0]
    that = []
    jstar = []
    tt = sp.Symbol('t')
    
    # Lagrange base polynomials
    polym1 = sp.Poly(((tt - 0   ) / (- dt - 0   ))*
                     ((tt -   dt) / (- dt -   dt))*
                     ((tt - 2*dt) / (- dt - 2*dt)))
    poly_0 = sp.Poly(((tt +   dt) / (0    +   dt))*
                     ((tt -   dt) / (0    -   dt))*
                     ((tt - 2*dt) / (0    - 2*dt)))
    polyp1 = sp.Poly(((tt +   dt) / (  dt +   dt))*
                     ((tt - 0   ) / (  dt - 0   ))*
                     ((tt - 2*dt) / (  dt - 2*dt)))
    polyp2 = sp.Poly(((tt +   dt) / (2*dt +   dt))*
                     ((tt - 0   ) / (2*dt - 0   ))*
                     ((tt -   dt) / (2*dt -   dt)))
    for j in jtemp:
        # since this is Lagrange interpolation, it does not work
        # for the first and last time steps, therefore the if
        if ((j > 1) and (j < y.shape[0]-2)):
            # define polynomial:
            poly = ((y[j-1] - yvalue)*polym1 +
                    (y[j  ] - yvalue)*poly_0 +
                    (y[j+1] - yvalue)*polyp1 +
                    (y[j+2] - yvalue)*polyp2)
            for poly_root in sp.nroots(poly):
                if poly_root.is_real:
                    if (poly_root > 0 and poly_root < 1):
                        that.append(poly_root)
                        jstar.append(j)
    xx = []
    zz = []
    for i in range(len(jstar)):
        j = jstar[i]
        tval = that[i]
        xx.append(x[j-1]*polym1.subs(tt, tval) +
                  x[j  ]*poly_0.subs(tt, tval) +
                  x[j+1]*polyp1.subs(tt, tval) +
                  x[j+2]*polyp2.subs(tt, tval))
        zz.append(z[j-1]*polym1.subs(tt, tval) +
                  z[j  ]*poly_0.subs(tt, tval) +
                  z[j+1]*polyp1.subs(tt, tval) +
                  z[j+2]*polyp2.subs(tt, tval))
    return xx, zz
```
