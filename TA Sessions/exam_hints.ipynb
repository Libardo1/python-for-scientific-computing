```python
import numpy as np
%matplotlib nbagg
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
def get_circle(
        radius = 1.,
        resolution = 8,
        orientation = 'z'):
    phi = np.linspace(0, 2*np.pi, num = resolution, endpoint = False)
    points = np.zeros((resolution, 3), dtype = np.float)
    p0 = radius*np.cos(phi)
    p1 = radius*np.sin(phi)
    lines = set([(i, (i+1) % resolution) for i in range(resolution)])
    if orientation == 'z':
        points[:, 0] = p0
        points[:, 1] = p1
    elif orientation == 'x':
        points[:, 1] = p0
        points[:, 2] = p1
    elif orientation == 'y':
        points[:, 2] = p0
        points[:, 0] = p1
    return points, lines

def get_sphere(
        radius = 1.,
        resolution = 8,
        center = np.zeros(3, dtype = np.float)):
    points = []
    lines = []
    z = np.linspace(-(radius - radius/resolution),
                     (radius - radius/resolution),
                     resolution)
    n = 0
    for zz in z:
        p, l = get_circle(radius = np.sqrt(radius**2 - zz**2),
                          resolution = resolution,
                          orientation = 'z')
        points.append(p + np.array([0., 0., zz]))
        lines += [[i[0]+n, i[1]+n] for i in l]
        n += resolution
    for i in range(resolution):
        for j in range(resolution-1):
            lines.append([i+j*resolution, i + (j+1)*resolution])
    return ((np.concatenate(points, axis = 0) +
             center),
            lines)

def plot_segments(a, points, connections):
    for c in connections:
        a.plot(points[list(c), 0], points[list(c), 1], points[list(c), 2], marker = '.')
    return None
```

```python
v, c = get_sphere(radius = 0.5)


fig = plt.figure()
a = fig.add_subplot(111, projection='3d')

n = 3
plot_segments(a, v, c)
t = np.linspace(0, 1, 2**10)
z = t - 0.5
rad = np.sqrt(0.25 - z**2)
x = np.cos(t*n*2*np.pi)*rad
y = np.sin(t*n*2*np.pi)*rad

a.plot(x, y, z, color = (0, 0, 0))
```
