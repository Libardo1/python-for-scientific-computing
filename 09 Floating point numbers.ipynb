Floating point numbers revisited
--------------------------------

```python
import numpy as np
%matplotlib nbagg
import matplotlib.pyplot as plt

sigma = 10.
beta  = 8./3
rho   = 28.

def lorenz_rhs(x):
    return np.array([sigma * (x[1] - x[0]),
                     x[0]*(rho - x[2]) - x[1],
                     x[0]*x[1] - beta*x[2]])

def Euler(rhs, dt = 0.5, N = 8, x0 = None):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    for t in range(N):
        x[t+1] = x[t] + dt*rhs(x[t])
    return x

def cRK(rhs, dt = 0.5, N = 8, x0 = None):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    for t in range(N):
        k1 = rhs(x[t])
        k2 = rhs(x[t] + 0.5*dt*k1)
        k3 = rhs(x[t] + 0.5*dt*k2)
        k4 = rhs(x[t] + dt*k3)
        x[t+1] = x[t] + dt*(k1 + 2*(k2 + k3) + k4)/6
    return x
```

```python
initial_condition = 10+np.random.random((3, 32))
lsol8 = cRK(lorenz_rhs,
             dt = 2.**(-7),
             N = 1024*2,
             x0 = initial_condition)
t8 = np.linspace(0, 2*1024*2.**(-7), 2*1024+1)
lsol7 = cRK(lorenz_rhs,
             dt = 2.**(-6),
             N = 512*2,
             x0 = initial_condition)

t7 = np.linspace(0, 2*512*2.**(-6), 2*512+1)
```

```python
ax = plt.figure().add_subplot(111)
ax.plot(t8, lsol8[:, :, 0])
ax.plot(t7, lsol7[:, :, 0])
#print np.average(np.abs(lsol8[-1] - lsol7[-1]), axis = (1))
```

**Conclusion:**

Even though we integrate the same trajectory (initial condition is the same), the use of a different time-step leads to "slow divergence" of the two numerical solutions.
This happens up to a time of about 9 (approximated from the plot).

After this time, deterministic chaos kicks in, and the trajectories become essentially distinct.
In the following, we show that we can interpret the distance between two trajectories with different timesteps as an error; this means that for systems which exhibit deterministic chaos, we cannot hope to obtain a "correct" numerical trajectory for indefinite time.

As a counter example, the ODE
\begin{equation}
    \frac{d}{dt} x = -x
\end{equation}
(and similarly whenever the initial condition for the ODE considered leads to an equilibrium point) can be solved numerically with any desired order of accuracy, because errors can be shown to decrease over time.

Errors for ODE integration methods
----------------------------------

```python
initial_condition = 10+np.random.random((3, 128))

epsilon = []
tstep = []
sol = []

N = 32
tstep.append(2.**(-3))
sol.append(cRK(lorenz_rhs,
               dt = tstep[-1],
               N = N,
               x0 = initial_condition))
for n in range(4, 9):
    tstep.append(2.**(-n))
    N = N*2
    sol.append(cRK(lorenz_rhs,
                   dt = tstep[-1],
                   N = N,
                   x0 = initial_condition))
    epsilon.append(sol[-1][-1] - sol[-2][-1])

tstep = np.array(tstep[:len(tstep)-1])
ax = plt.figure().add_subplot(111)
ax.plot(tstep, np.average(np.abs(epsilon), axis = (1, 2)), marker = '.')
ax.plot(tstep, 1e4*tstep**4, marker = '.')
ax.set_xscale('log')
ax.set_yscale('log')
```

```python
def get_evdt(
        initial_condition,
        h0, N0, ndivisions,
        method,
        rhs):
    epsilon = []
    tstep = [h0]
    sol = []

    N = N0
    sol.append(method(rhs,
                      dt = tstep[-1],
                      N = N,
                      x0 = initial_condition))
    for n in range(1, ndivisions + 1):
        tstep.append(h0*2.**(-n))
        N = N*2
        sol.append(method(rhs,
                          dt = tstep[-1],
                          N = N,
                          x0 = initial_condition))
        epsilon.append(sol[-1][-1] - sol[-2][-1])

    return (np.array(tstep[:len(tstep)-1]),
            np.abs(epsilon))
```

```python
ic = 10+np.random.random((3, 128))

Edt, Ee = get_evdt(ic, 2.**(-5), 32, 9, Euler, lorenz_rhs)
cRKdt, cRKe = get_evdt(ic, 2.**(-5), 32, 11, cRK, lorenz_rhs)
```

```python
ax = plt.figure().add_subplot(111)
ax.plot(Edt, np.average(Ee, axis = (1, 2)), marker = '.', label = 'Euler')
ax.plot(cRKdt, np.average(cRKe, axis = (1, 2)), marker = '.', label = 'cRK')
ax.plot(Edt, 1e3*Edt, label = '$\Delta t$')
ax.plot(Edt, 1e4*(Edt**4), label = '$\Delta t^4$')
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend(loc = 'best')
```

```python
ic = 10+np.random.random((3, 128))

dt1, e1 = get_evdt(ic.astype(np.float32), 2.**(-7), 32, 9, cRK, lorenz_rhs)
dt2, e2 = get_evdt(ic.astype(np.float64), 2.**(-7), 32, 9, cRK, lorenz_rhs)
dt4, e4 = get_evdt(ic.astype(np.float128), 2.**(-7), 32, 9, cRK, lorenz_rhs)

ax = plt.figure().add_subplot(111)
ax.plot(dt1, np.average(e1, axis = (1, 2)), marker = '.', label = '1')
ax.plot(dt2, np.average(e2, axis = (1, 2)), marker = '.', label = '2')
ax.plot(dt4, np.average(e4, axis = (1, 2)), marker = '.', label = '4')
ax.plot(dt2, 1e4*(dt2**4), label = '$\Delta t^4$')
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend(loc = 'best')
```

```python
def cRK_Kahan(rhs, dt = 0.5, N = 8, x0 = None):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    c = 0.0
    for t in range(N):
        k1 = rhs(x[t])
        k2 = rhs(x[t] + 0.5*dt*k1)
        k3 = rhs(x[t] + 0.5*dt*k2)
        k4 = rhs(x[t] + dt*k3)
        n = (k1 + 2*(k2 + k3) + k4)*dt/6
        y = n - c
        T = x[t] + y
        c = (T - x[t]) - y
        x[t+1] = T
    return x
```

```python
dt1k, e1k = get_evdt(ic.astype(np.float32), 2.**(-7), 32, 9, cRK_Kahan, lorenz_rhs)

ax = plt.figure().add_subplot(111)
ax.plot(dt1, np.average(e1, axis = (1, 2)), marker = '.', label = '1')
ax.plot(dt1, np.average(e1k, axis = (1, 2)), marker = '.', label = '1k')
ax.plot(dt2, np.average(e2, axis = (1, 2)), marker = '.', label = '2')
ax.plot(dt4, np.average(e4, axis = (1, 2)), marker = '.', label = '4')
ax.plot(dt2, 1e4*(dt2**4), label = '$\Delta t^4$')
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend(loc = 'best')
```

```python
import numpy as np
%matplotlib nbagg
import matplotlib.pyplot as plt

def Mandelbrot(xmin, xmax, nx, ymin, ymax, ny, max_level, dtype = np.float32):
    x = np.linspace(xmin, xmax, nx).astype(dtype)
    y = np.linspace(ymax, ymin, ny).astype(dtype)
    c = x[np.newaxis, :] + 1j*y[:, np.newaxis]
    print c.dtype
    z = np.zeros(c.shape, c.dtype)
    iterations = np.zeros(z.shape, np.int)
    for n in range(max_level):
        good_indices = np.where(np.abs(z) < 2)
        z[good_indices] = z[good_indices]**2 + c[good_indices]
        iterations[good_indices] += 1
    ax = plt.figure(figsize=(3,3)).add_axes([.0, .0, 1., 1.])
    ax.set_axis_off()
    ax.imshow(iterations, interpolation = 'none')
    return x, y, iterations

def Mandelbrot_zoom(Z = complex(-1.6906816638949, 0.0),
                    factor = 1e-4,
                    N = 200,
                    dtype = np.float32,
                    resolution = 256):
    xm = Z.real - factor
    xM = Z.real + factor
    ym = Z.imag - factor
    yM = Z.imag + factor
    Mandelbrot(xm, xM, resolution,
               ym, yM, resolution,
               N, dtype = dtype)
    
Mandelbrot_zoom(dtype = np.float32)
Mandelbrot_zoom(dtype = np.float64)
Mandelbrot_zoom(dtype = np.float128)
```

```python

p = 7
for dt in [np.float32, np.float64, np.float128]:
    Mandelbrot_zoom(factor = 10.**(-p),
                    dtype = dt)
```
