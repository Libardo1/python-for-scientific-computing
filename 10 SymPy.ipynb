```python
def centered_differences(neighbours = 1):
    npoints = neighbours*2+1
    x = sp.Symbol('x')
    f = [sp.Symbol('f_{0}'.format(i)) for i in range(npoints)]
    t = sum(f[i]*x**i/sp.factorial(i) for i in range(len(f)))
    h = sp.Symbol('h')
    F = [sp.Symbol('F({0})'.format(i*h)) for i in range(-neighbours, neighbours+1)]
    return sp.solve([t.subs(x, -i*h) - F[i+neighbours]
                     for i in range(-neighbours, neighbours+1)],
                    f)

print centered_differences(1)
print centered_differences(2)
print sp.latex(centered_differences(1))
print sp.latex(centered_differences(2))
```

Newton method
-------------

\begin{equation}
    x_n = x_{n-1} - \frac{f(x_{n-1})}{f_x(x_{n-1})}
\end{equation}

```python
x = sp.Symbol('x')
f = x**3 - 2*x**2 + 4

def Newton(var = None,
           function = None,
           x0 = 0.0,
           max_loop = 20,
           max_err = 1e-6):
    err = 1.0
    i = 0
    while (err > max_err and
           i < max_loop):
        x0 = x0 - function.subs(var, x0) / function.diff(var).subs(var, x0)
        err = abs(function.subs(var, x0))
        i += 1
    return x0

print Newton(var = x, function = f,
             x0 = 1.0, max_err = 1e-8)
```

Taylor ODE methods
------------------

```python
import numpy as np
%matplotlib nbagg
import matplotlib.pyplot as plt

sigma = 10.
beta  = 8./3
rho   = 28.

def lorenz_rhs(x):
    return np.array([sigma * (x[1] - x[0]),
                     x[0]*(rho - x[2]) - x[1],
                     x[0]*x[1] - beta*x[2]])

def Euler(rhs, dt = 0.5, N = 8, x0 = None):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    for t in range(N):
        x[t+1] = x[t] + dt*rhs(x[t])
    return x

def cRK(rhs, dt = 0.5, N = 8, x0 = None):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    for t in range(N):
        k1 = rhs(x[t])
        k2 = rhs(x[t] + 0.5*dt*k1)
        k3 = rhs(x[t] + 0.5*dt*k2)
        k4 = rhs(x[t] + dt*k3)
        x[t+1] = x[t] + dt*(k1 + 2*(k2 + k3) + k4)/6
    return x

def get_evdt(
        initial_condition,
        h0, N0, ndivisions,
        method,
        rhs):
    epsilon = []
    tstep = [h0]
    sol = []

    N = N0
    sol.append(method(rhs,
                      dt = tstep[-1],
                      N = N,
                      x0 = initial_condition))
    for n in range(1, ndivisions + 1):
        tstep.append(h0*2.**(-n))
        N = N*2
        sol.append(method(rhs,
                          dt = tstep[-1],
                          N = N,
                          x0 = initial_condition))
        epsilon.append(sol[-1][-1] - sol[-2][-1])

    return (np.array(tstep[:len(tstep)-1]),
            np.abs(epsilon))
```

```python
ic = 10+np.random.random((3, 128))

Edt, Ee = get_evdt(ic, 2.**(-5), 32, 9, Euler, lorenz_rhs)
cRKdt, cRKe = get_evdt(ic, 2.**(-5), 32, 7, cRK, lorenz_rhs)
```

```python
def Taylor2(varlist, rhslist, dt = 0.5, N = 8, x0 = None):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    difflist = [sum(rhslist[i]*rhs.diff(varlist[i]) for i in range(len(varlist)))
                for rhs in rhslist]
    vel = [sp.utilities.lambdify(tuple(varlist), rhs)
           for rhs in rhslist]
    acc = [sp.utilities.lambdify(tuple(varlist), rhs)
           for rhs in difflist]
    for t in range(N):
        for k in range(x.shape[1]):
            x[t+1, k] = x[t, k] + dt*vel[k](*tuple(x[t])) + dt**2 * acc[k](*tuple(x[t])) / 2
    return x
```

```python
def get_evdt_sp(
        initial_condition,
        h0, N0, ndivisions,
        method,
        var,
        rhs):
    epsilon = []
    tstep = [h0]
    sol = []

    N = N0
    sol.append(method(var, rhs,
                      dt = tstep[-1],
                      N = N,
                      x0 = initial_condition))
    for n in range(1, ndivisions + 1):
        tstep.append(h0*2.**(-n))
        N = N*2
        sol.append(method(var, rhs,
                          dt = tstep[-1],
                          N = N,
                          x0 = initial_condition))
        epsilon.append(sol[-1][-1] - sol[-2][-1])

    return (np.array(tstep[:len(tstep)-1]),
            np.abs(epsilon))

x = sp.Symbol('x')
y = sp.Symbol('y')
z = sp.Symbol('z')

sigma = 10.
beta  = 8./3
rho   = 28.

lorenz_func = [sigma * (y - x),
               x*(rho - z) - y,
               x*y - beta*z]


T2dt, T2e = get_evdt_sp(ic, 2.**(-5), 32, 9, Taylor2, [x, y, z], lorenz_func)
```

```python
def GRK2(rhs, dt = 0.5, N = 8, x0 = None, err = 1e-6):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    y = x0.copy()
    z = x0.copy()
    for t in range(N):
        # begin solution of implicit equation
        y = x[t] + .5*dt*rhs(x[t])
        z = x[t] + .5*dt*rhs(y)
        # use fixed point iteration to find solution
        while np.average(np.abs(y - z) /
                         np.max(np.abs(z))) > err:
            y = z.copy()
            z = x[t] + .5*dt*rhs(y)
        # end solution of implicit equation
        x[t+1] = x[t] + dt*rhs(z)
    return x


GRK2dt, GRK2e = get_evdt(ic, 2.**(-5), 32, 9, GRK2, lorenz_rhs)
```

The 2nd order Gauss Runge Kutta (or midpoint) method is, as the name specifies, a second order method (same as Taylor 2).
It is an implicit method, in the sense that at each time step an implicit equation must be solved.
The implicit equation is as follows:
\begin{equation}
    \mathbf{y} = \mathbf{x}(t) + \frac{h}{2} \mathbf{f}(\mathbf{y})
\end{equation}
Using the solution of this equation, we can approximate $\mathbf{x}(t+h)$ as:
\begin{equation}
    \mathbf{x}(t+h) \approx \mathbf{x}(t) + h \mathbf{f}(\mathbf{y})
\end{equation}

In order to solve the equation for $\mathbf{y}$, we use fixed point iteration: https://en.wikipedia.org/wiki/Fixed-point_iteration.
We define a function $\mathbf{F}$:
\begin{equation}
    \mathbf{F}(\mathbf{y}) \equiv
    \mathbf{x}(t) + \frac{h}{2} \mathbf{f}(\mathbf{y})
\end{equation}
and then we look for its fixed point (`while` loop from the code).

Obviously, we should be a little worried about how many iterations are required, since evaluating the right hand side can be quite expensive in certain scenarios.

```python
def GRK2_iter(rhs, dt = 0.5, N = 8, x0 = None, err = 1e-15):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    y = x0.copy()
    z = x0.copy()
    iterations = []
    for t in range(N):
        y = x[t] + .5*dt*rhs(x[t])
        z = x[t] + .5*dt*rhs(y)
        i = 0
        while ((np.average(np.abs(y - z) /
                           np.max(np.abs(z))) > err) and
                i < 20):
            y = z.copy()
            z = x[t] + .5*dt*rhs(y)
            i += 1
        iterations.append(i)
        x[t+1] = x[t] + dt*rhs(z)
    print('average number of iterations for dt {0} was {1}'.format(dt, np.average(iterations)))
    return x

GRK2dt, GRK2e = get_evdt(ic, 2.**(-5), 32, 7, GRK2_iter, lorenz_rhs)
```

```python
def GRK2_Newton(varlist, rhslist, dt = 0.5, N = 8, x0 = None, err = 1e-15):
    dim = x0.shape[0]
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    y = x0.copy()
    z = x0.copy()
    
    vel = [sp.utilities.lambdify(tuple(varlist), rhs)
           for rhs in rhslist]
    
    Jsymb = [[sp.utilities.lambdify(tuple(varlist), rhs.diff(var))
              for var in varlist]
             for rhs in rhslist]
    def f(point):
        return np.array([vel[k](*tuple(point))
                         for k in range(dim)])
    def J(point):
        return np.array([np.array([(Jsymb[i][j](*tuple(point)) *
                                    np.ones(point.shape[1:], point.dtype))
                                   for j in range(dim)])
                         for i in range(dim)])
    
    iterations = []
    for t in range(N):
        y = x[t] + .5*dt*f(x[t])
        z = x[t] + .5*dt*f(y)
        i = 0
        while ((np.average(np.abs(y - z) /
                           np.max(np.abs(z))) > err) and
                i < 20):
            y = z.copy()
            Jval = J(y).transpose(2, 0, 1)
            dy = np.linalg.solve(.5*dt*Jval,
                                 -(x[t] + .5*dt*f(y)).transpose())
            
            z = y + dy.transpose()
            i += 1
        iterations.append(i)
        x[t+1] = x[t] + dt*f(z)
    print('average number of iterations was {0}'.format(np.average(iterations)))
    return x


x = sp.Symbol('x')
y = sp.Symbol('y')
z = sp.Symbol('z')

sigma = 10.
beta  = 8./3
rho   = 28.

lorenz_func = [sigma * (y - x),
               x*(rho - z) - y,
               x*y - beta*z]

GRK2spdt, GRK2spe = get_evdt_sp(ic, 2.**(-5), 32, 6,
                                GRK2_Newton,
                                [x, y, z],
                                lorenz_func)
print(np.average(GRK2spe, axis = (1, 2)))
```
