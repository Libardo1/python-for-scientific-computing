```python
# general case
import sympy as sp

x = sp.Symbol('x')
f = [sp.Symbol('f{0}'.format(i)) for i in range(3)]
t = f[0] + f[1]*x + (f[2]*x**2)/2
F = [sp.Symbol('F({0}*h)'.format(i)) for i in range(-1, 2)]
h = sp.Symbol('h')

sol = sp.solve(
        [t.subs(x, -h) - F[0],
         t.subs(x, 0) - F[1],
         t.subs(x, h) - F[2]],
        f)

print(sol)
```

```python
# general case
import sympy as sp

x = sp.Symbol('x')
f = [sp.Symbol('f{0}'.format(i)) for i in range(5)]
t = sum(f[i]*x**i/sp.factorial(i) for i in range(len(f)))
F = [sp.Symbol('F({0}*h)'.format(i)) for i in range(-2, 3)]
h = sp.Symbol('h')

sol = sp.solve([t.subs(x, -i*h) - F[i+2]
                for i in range(-2, 3)],
               f)

print(sol)
```

```python
def centered_differences(neighbours = 1):
    npoints = neighbours*2+1
    x = sp.Symbol('x')
    f = [sp.Symbol('f_{0}'.format(i)) for i in range(npoints)]
    t = sum(f[i]*x**i/sp.factorial(i) for i in range(len(f)))
    h = sp.Symbol('h')
    F = [sp.Symbol('F({0})'.format(i*h)) for i in range(-neighbours, neighbours+1)]
    return sp.solve([t.subs(x, -i*h) - F[i+neighbours]
                     for i in range(-neighbours, neighbours+1)],
                    f)

print centered_differences(1)
print centered_differences(2)
print sp.latex(centered_differences(1))
print sp.latex(centered_differences(2))
```

\begin{equation}
    \left \{ f_{0} : F(0), \quad f_{1} : \frac{F(-h) - F(h)}{2 h}, \quad f_{2} : \frac{1}{h^{2}} \left(F(-h) - 2 F(0) + F(h)\right)\right \}
\end{equation}


\begin{align}
    f_{0} &: F(0), \\
    f_{1} &: \frac{1}{12 h} \left(- F(-2h) + 8 F(-h) + F(2h) - 8 F(h)\right), \\
    f_{2} &: \frac{1}{12 h^{2}} \left(- F(-2h) + 16 F(-h) - 30 F(0) - F(2h) + 16 F(h)\right), \\
    f_{3} &: \frac{1}{h^{3}} \left(\frac{F(-2h)}{2} - F(-h) - \frac{F(2h)}{2} + F(h)\right), \\
    f_{4} &: \frac{1}{h^{4}} \left(F(-2h) - 4 F(-h) + 6 F(0) + F(2h) - 4 F(h)\right)
\end{align}

Newton method
-------------

\begin{equation}
    x_n = x_{n-1} - \frac{f(x_{n-1})}{f_x(x_{n-1})}
\end{equation}

```python
x = sp.Symbol('x')
f = x**3 - 2*x**2 + 4

def Newton(var = None,
           function = None,
           x0 = 0.0,
           max_loop = 20,
           max_err = 1e-6):
    err = 1.0
    i = 0
    while (err > max_err and
           i < max_loop):
        x0 = x0 - function.subs(var, x0) / function.diff(var).subs(var, x0)
        err = abs(function.subs(var, x0))
        i += 1
    return x0

print Newton(var = x, function = f,
             x0 = 1.0, max_err = 1e-8)
```

Taylor ODE methods
------------------

```python
import numpy as np
%matplotlib nbagg
import matplotlib.pyplot as plt

sigma = 10.
beta  = 8./3
rho   = 28.

def lorenz_rhs(x):
    return np.array([sigma * (x[1] - x[0]),
                     x[0]*(rho - x[2]) - x[1],
                     x[0]*x[1] - beta*x[2]])

def Euler(rhs, dt = 0.5, N = 8, x0 = None):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    for t in range(N):
        x[t+1] = x[t] + dt*rhs(x[t])
    return x

def cRK(rhs, dt = 0.5, N = 8, x0 = None):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    for t in range(N):
        k1 = rhs(x[t])
        k2 = rhs(x[t] + 0.5*dt*k1)
        k3 = rhs(x[t] + 0.5*dt*k2)
        k4 = rhs(x[t] + dt*k3)
        x[t+1] = x[t] + dt*(k1 + 2*(k2 + k3) + k4)/6
    return x

def get_evdt(
        initial_condition,
        h0, N0, ndivisions,
        method,
        rhs):
    epsilon = []
    tstep = [h0]
    sol = []

    N = N0
    sol.append(method(rhs,
                      dt = tstep[-1],
                      N = N,
                      x0 = initial_condition))
    for n in range(1, ndivisions + 1):
        tstep.append(h0*2.**(-n))
        N = N*2
        sol.append(method(rhs,
                          dt = tstep[-1],
                          N = N,
                          x0 = initial_condition))
        epsilon.append(sol[-1][-1] - sol[-2][-1])

    return (np.array(tstep[:len(tstep)-1]),
            np.abs(epsilon))
```

```python
ic = 10+np.random.random((3, 128))

Edt, Ee = get_evdt(ic, 2.**(-5), 32, 9, Euler, lorenz_rhs)
cRKdt, cRKe = get_evdt(ic, 2.**(-5), 32, 7, cRK, lorenz_rhs)
```

```python
ax = plt.figure().add_subplot(111)
ax.plot(Edt, np.average(Ee, axis = (1, 2)), marker = '.', label = 'Euler')
ax.plot(cRKdt, np.average(cRKe, axis = (1, 2)), marker = '.', label = 'cRK')
ax.plot(Edt, 1e3*Edt, label = '$\Delta t$')
ax.plot(Edt, 1e4*(Edt**4), label = '$\Delta t^4$')
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend(loc = 'best')
```

```python
def Taylor2(varlist, rhslist, dt = 0.5, N = 8, x0 = None):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    difflist = [sum(rhslist[i]*rhs.diff(varlist[i]) for i in range(len(varlist)))
                for rhs in rhslist]
    vel = [sp.utilities.lambdify(tuple(varlist), rhs)
           for rhs in rhslist]
    acc = [sp.utilities.lambdify(tuple(varlist), rhs)
           for rhs in difflist]
    for t in range(N):
        for k in range(x.shape[1]):
            x[t+1, k] = x[t, k] + dt*vel[k](*tuple(x[t])) + dt**2 * acc[k](*tuple(x[t])) / 2
    return x
```

```python
def get_evdt_sp(
        initial_condition,
        h0, N0, ndivisions,
        method,
        var,
        rhs):
    epsilon = []
    tstep = [h0]
    sol = []

    N = N0
    sol.append(method(var, rhs,
                      dt = tstep[-1],
                      N = N,
                      x0 = initial_condition))
    for n in range(1, ndivisions + 1):
        tstep.append(h0*2.**(-n))
        N = N*2
        sol.append(method(var, rhs,
                          dt = tstep[-1],
                          N = N,
                          x0 = initial_condition))
        epsilon.append(sol[-1][-1] - sol[-2][-1])

    return (np.array(tstep[:len(tstep)-1]),
            np.abs(epsilon))

x = sp.Symbol('x')
y = sp.Symbol('y')
z = sp.Symbol('z')

sigma = 10.
beta  = 8./3
rho   = 28.

lorenz_func = [sigma * (y - x),
               x*(rho - z) - y,
               x*y - beta*z]


T2dt, T2e = get_evdt_sp(ic, 2.**(-5), 32, 9, Taylor2, [x, y, z], lorenz_func)
```

```python
ax = plt.figure().add_subplot(111)
ax.plot(Edt, np.average(Ee, axis = (1, 2)), marker = '.', label = 'Euler')
ax.plot(T2dt, np.average(T2e, axis = (1, 2)), marker = '.', label = 'T2')
ax.plot(cRKdt, np.average(cRKe, axis = (1, 2)), marker = '.', label = 'cRK')
ax.plot(Edt, 1e3*Edt, label = '$\Delta t$')
ax.plot(Edt, 1e4*(Edt**4), label = '$\Delta t^4$')
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend(loc = 'best')
```
