Symbolic computing
------------------

The point is to use objects that "know" exact rules of mathematics.
Function objects, number objects, variable objects, etc.

```python
x = sp.Symbol('x')
f = x*3 + sp.exp(x) - x**3.75

print f.diff(x)
```

```python
x = sp.Symbol('y')
f = x*3 + sp.exp(x) - x**3.75

print f.diff(x)
```

```python
from sympy import latex
print(latex(f.diff(x)))
```

Finite differences
------------------

```python
import numpy as np
%matplotlib nbagg
import matplotlib.pyplot as plt

t = np.linspace(0, 2*np.pi, 16)
a = plt.figure().add_subplot(111)
y = np.sin(t) + 0.25*np.cos(3.5*t)
ndy = (y[1:] - y[:y.shape[0]-1]) / (t[1:] - t[:t.shape[0]-1])
dy = np.cos(t) - 0.25*3.5*np.sin(3.5*t)
a.plot(t, y, marker = '.', label = 'y')
a.plot(t[:t.shape[0]-1], ndy, marker = '.', label = 'dnumerical')
a.plot(t, dy, marker = '.', label = 'dexact')
a.legend(loc = 'best')
```

```python
t = np.linspace(0, 2*np.pi, 32)
a = plt.figure().add_subplot(111)
y = np.sin(t) + 0.25*np.cos(3.5*t)
ndy = (y[1:] - y[:y.shape[0]-1]) / (t[1:] - t[:t.shape[0]-1])
dy = np.cos(t) - 0.25*3.5*np.sin(3.5*t)
a.plot(t, y, marker = '.', label = 'y')
a.plot(t, np.zeros(t.shape), marker = '.', label = 'nothing', color = (0, 0, 0))
a.plot(t[:t.shape[0]-1], ndy, marker = '.', label = 'dnumerical')
a.plot(t, dy, marker = '.', label = 'dexact')
a.legend(loc = 'best')
```

```python
x = sp.Symbol('x')
f = [sp.Symbol('f{0}'.format(i)) for i in range(3)]
t = f[0] + f[1]*x + (f[2]*x**2)/2
F = [sp.Symbol('F({0})'.format(i)) for i in range(-1, 2)]

sol = sp.solve(
        [t.subs(x, -1) - F[0],
         t.subs(x, 0) - F[1],
         t.subs(x, 1) - F[2]],
        f)
print sol
```

```python
t = np.linspace(0, 2*np.pi, 32)
a = plt.figure().add_subplot(111)
y = np.sin(t) + 0.25*np.cos(3.5*t)
ndy = (y[2:] - y[:y.shape[0]-2]) / ((t[2:] - t[:t.shape[0]-2]))

ndyold = (y[1:] - y[:y.shape[0]-1]) / (t[1:] - t[:t.shape[0]-1])
dy = np.cos(t) - 0.25*3.5*np.sin(3.5*t)
a.plot(t, y, marker = '.', label = 'y')
a.plot(t, np.zeros(t.shape), marker = '.', label = 'nothing', color = (0, 0, 0))
a.plot(t[1:t.shape[0]-1], ndy, marker = '.', label = 'dnumerical')
a.plot(t[:t.shape[0]-1], ndyold, marker = '.', label = 'dnumericalold')
a.plot(t, dy, marker = '.', label = 'dexact')
a.legend(loc = 'best')
```
