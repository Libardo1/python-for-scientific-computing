\begin{equation}
    f_c: \mathbb{C} \rightarrow {\mathbb{C}}, f_c(z) = z^2 + c
\end{equation}


we are interested in the bounded set $M$ such that $f_c^n(0) \in M$ ($f_c$ composed with itself $n$ times) for all $n \in \mathbb{N}$, for all $c \in M$.

```python
import numpy as np

x = np.linspace(-2, 2, 256)
y = np.linspace(2, 0, 128)

z = x[np.newaxis, :] + 1j*y[:, np.newaxis]
iterations = np.zeros(z.shape, np.int)

def quadratic_map(z, c = 0):
    return z**2 + c

for j in range(y.shape[0]):
    for i in range(x.shape[0]):
        z0 = 0
        nmax = 0
        for n in range(1, 20):
            if abs(z0) > 2:
                break
            z0 = quadratic_map(z0, z[j, i])
            nmax = n
        iterations[j, i] = nmax
```

```python
%matplotlib nbagg
import matplotlib.pyplot as plt

ax = plt.figure().add_subplot(211)
ax.imshow(iterations, interpolation = 'none', extent = [x[0], x[-1], y[-1], y[0]])

iter2 = np.zeros(iterations.shape, iterations.dtype)
indices = []
for i in range(10):
    indices.append(np.where(iterations == i))
iter2[indices[6]] = 1

ax = plt.gcf().add_subplot(212)
ax.imshow(iter2, interpolation = 'none', extent = [x[0], x[-1], y[-1], y[0]])
```

Boring stuff
------------

we are going to identify clusters of points.

we look at our data as a graph, in the sense that each point is a vertex, and it is connected to immediate grid neighbours.

we then pick out subsets of vertices (with `np.where`), and we try to find connected subgraphs for the graphs made from these subsets.

```python

class halfM:
    def __init__(self, N):
        self.N = N
        self.x = np.linspace(-2, 2, 2*self.N)
        self.y = np.linspace(2, 0, self.N)
        self.points = self.x[np.newaxis, :] + 1j*self.y[:, np.newaxis]
        self.iterations = np.zeros((self.y.shape[0], self.x.shape[0]), np.int)
        return None
    def __call__(self, n):
        c = self.x[np.newaxis, :] + 1j*self.y[:, np.newaxis]
        z = np.zeros(c.shape, c.dtype)
        self.iterations[:] = 0
        for i in range(1, n):
            good_indices = np.where(np.abs(z) < 2)
            z[good_indices] = quadratic_map(z[good_indices], c[good_indices])
            self.iterations[good_indices] += 1
        return None
    def get_clusters(self, n):
        good_indices = np.where(self.iterations == n)
        return None
        
bla = halfM(64)
bla(20)
bla.get_clusters(7)
print(bla.iterations[good_indices].shape)
print(bla.iterations[good_indices[0], good_indices[1]].shape)
print(np.where(np.logical_not(
        bla.iterations[good_indices] == bla.iterations[good_indices[0], good_indices[1]])))
```

```python
class halfM:
    def __init__(self, N):
        self.N = N
        self.x = np.linspace(-2, 2, 2*self.N)
        self.y = np.linspace(2, 0, self.N)
        self.points = self.x[np.newaxis, :] + 1j*self.y[:, np.newaxis]
        self.iterations = np.zeros((self.y.shape[0], self.x.shape[0]), np.int)
        return None
    def __call__(self, n):
        c = self.x[np.newaxis, :] + 1j*self.y[:, np.newaxis]
        z = np.zeros(c.shape, c.dtype)
        self.max_level = n-1
        self.iterations[:] = 0
        for i in range(1, n):
            good_indices = np.where(np.abs(z) <= 2)
            z[good_indices] = quadratic_map(z[good_indices], c[good_indices])
            self.iterations[good_indices] += 1
        return None
    def get_neighbours(self, yi, xi):
        n = []
        for p in [(yi  , xi-1),
                  (yi-1, xi  ),
                  (yi-1, xi-1),
                  (yi  , xi+1),
                  (yi+1, xi  ),
                  (yi+1, xi+1),
                  (yi+1, xi-1),
                  (yi-1, xi+1)]:
            if ((p[0] >=0 and p[0] < self.iterations.shape[0]) and
                (p[1] >=0 and p[1] < self.iterations.shape[1])):
                n.append(p)
        return n
    def get_clusters(self, n):
        yi, xi = np.where(self.iterations == n)
        all_points = set([(yi[i], xi[i]) for i in range(yi.shape[0])])
        cluster = []
        while not len(all_points) == 0:
            point = all_points.pop()
            cluster.append(set())
            to_check = [point]
            while not (len(to_check) == 0):
                point = to_check.pop()
                if (self.iterations[point[0], point[1]] == n and
                    (point not in cluster[-1])):
                    for p in self.get_neighbours(point[0], point[1]):
                        to_check.append(p)
                    cluster[-1].add(point)
                    if (point in all_points):
                        all_points.remove(point)
        return cluster
    def get_edge(self):
        n = self.max_level
        cluster = []
        point = (self.iterations.shape[0]-1, 0)
        to_check = [point]
        while not (len(to_check) == 0):
            point = to_check.pop()
            if (self.iterations[point[0], point[1]] == n and
                (point not in cluster)):
                good_point = False
                for p in self.get_neighbours(point[0], point[1]):
                    if self.iterations[p[0], p[1]] < n:
                        good_point = True
                    else:
                        needs_appending = False
                        for pp in self.get_neighbours(p[0], p[1]):
                            if self.iterations[pp[0], pp[1]] < n:
                                needs_appending = True
                        if needs_appending:
                            to_check.append(p)
                if good_point:
                    cluster.append(point)
        return cluster
    
bla = halfM(64)
bla(20)
c = bla.get_edge()
ax = plt.figure().add_subplot(111)
ax.set_axis_off()
iterations = bla.iterations.copy()
iterations[:] = 0

actual_points = []
for cc in c:
    iterations[cc[0], cc[1]] = 1
    actual_points.append(bla.points[cc[0], cc[1]])
ax.imshow(iterations, interpolation = 'none')

actual_points = np.array(actual_points)

ax = plt.figure().add_subplot(111)
ax.plot(np.real(actual_points), np.imag(actual_points))
```
