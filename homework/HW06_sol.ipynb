A. Earth - Moon
---------------

In the following, the code developed in class is used to simulate the Earth-Moon system (not the exact system, since the velocity and distance are only approximate, but it should work for a few "months").

Please make an animation of the system, similarly to what was done for the double-pendulum system for instance.

```python
%matplotlib nbagg
import matplotlib.pyplot as plt
from matplotlib import animation
import sympy as sp
import numpy as np
```

```python
class hamiltonian_system:
    def __init__(self, x, p, H):
        assert(len(x) == len(p))
        self.H = H
        self.x = x
        self.p = p
        self.degrees_of_freedom = len(self.x)
        self.rhs = ([sp.utilities.lambdify(tuple(self.x)+tuple(self.p),
                                           self.H.diff(self.p[i]),
                                           modules = 'numpy')
                     for i in range(self.degrees_of_freedom)] +
                    [sp.utilities.lambdify(tuple(self.x)+tuple(self.p),
                                           -self.H.diff(self.x[i]),
                                           modules = 'numpy')
                     for i in range(self.degrees_of_freedom)])
        self.energy = sp.utilities.lambdify(tuple(self.x)+tuple(self.p),
                                            self.H,
                                            modules = 'numpy')
        return None
    def numpy_rhs(self, point):
        return np.array([self.rhs[i](*tuple(point)) * np.ones(point.shape[1:])
                         for i in range(2*self.degrees_of_freedom)])
```

```python
def Euler(rhs, dt = 0.5, N = 8, x0 = None):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    for t in range(N):
        x[t+1] = x[t] + dt*rhs(x[t])
    return x

def cRK(rhs, dt = 0.5, N = 8, x0 = None):
    x = np.zeros((N+1,) + x0.shape,
                 dtype = x0.dtype)
    x[0] = x0
    for t in range(N):
        k1 = rhs(x[t])
        k2 = rhs(x[t] + 0.5*dt*k1)
        k3 = rhs(x[t] + 0.5*dt*k2)
        k4 = rhs(x[t] + dt*k3)
        x[t+1] = x[t] + dt*(k1 + 2*(k2 + k3) + k4)/6
    return x

def get_evdt(
        initial_condition,
        h0, N0, ndivisions,
        method,
        rhs):
    epsilon = []
    tstep = [h0]
    sol = []

    N = N0
    sol.append(method(rhs,
                      dt = tstep[-1],
                      N = N,
                      x0 = initial_condition))
    for n in range(1, ndivisions + 1):
        tstep.append(h0*2.**(-n))
        N = N*2
        sol.append(method(rhs,
                          dt = tstep[-1],
                          N = N,
                          x0 = initial_condition))
        epsilon.append((sol[-1][-1] - sol[-2][-1]) /
                       np.max(np.abs(sol[-1])))

    return (np.array(tstep[:len(tstep)-1]),
            np.abs(epsilon))
```

```python
r1x = sp.Symbol('r1x')
r1y = sp.Symbol('r1y')
r2x = sp.Symbol('r2x')
r2y = sp.Symbol('r2y')

p1x = sp.Symbol('p1x')
p1y = sp.Symbol('p1y')
p2x = sp.Symbol('p2x')
p2y = sp.Symbol('p2y')

G  = 6.67384e-11
m1 = 5.97219e24
m2 = 7.34770e22

H = ((p1x**2 + p1y**2)/(2*m1) + (p2x**2 + p2y**2)/(2*m2)
     -(G*m1*m2)/sp.sqrt((r1x - r2x)**2 + (r1y - r2y)**2))

earth_and_moon = hamiltonian_system([r1x, r1y, r2x, r2y],
                                    [p1x, p1y, p2x, p2y],
                                    H)

vel = 1.022e3
dist = 386.0e6

# the variables in Hamilton's equations are positions and momenta,
# therefore we have vel*m2 in the initial condition
initial_condition = np.array([0., 0., dist, 0.,
                              0., 0., 0., vel*m2])
```

```python
plt.close('all')
dt, err = get_evdt(initial_condition,
                   2.**(12), 32, 8,
                   cRK,
                   earth_and_moon.numpy_rhs)
fig = plt.figure()
a = fig.add_subplot(111)
a.plot(dt, np.average(err, axis = 1), marker = '.')
a.set_xscale('log')
a.set_yscale('log')
```

```python
sol1 = cRK(earth_and_moon.numpy_rhs,
           dt = 2.**(10),
           N = 2**12,
           x0 = initial_condition)
```

```python
# total momentum has to be conserved, therefore
# the center of mass for the 2 particle system is moving "up"
fig = plt.figure(figsize=(6,6))
a = fig.add_subplot(111)
a.plot(sol1[:, 0], sol1[:, 1])
a.plot(sol1[:, 2], sol1[:, 3])
```

```python
# an initial condition for which the total
# momentum is 0
# vcm = (p1 + p2)/(m1+m2)
# vcm = 0
# v2 - v1 = vel
# p1 = -p2
# m1*v1 = -m2*v2
# m1*(vel + v2) = -m2*v2
# m1*vel = -(m2+m1)*v2
# v2 = vel*m1/(m2+m1)
# p1 = vel*m1*m2/(m1+m2)
P = vel*m1*m2/(m1+m2)
initial_condition = np.array([0., 0., dist, 0.,
                              0., -P, 0., P])
sol2 = cRK(earth_and_moon.numpy_rhs,
           dt = 2.**(10),
           N = 2**12,
           x0 = initial_condition)


fig = plt.figure(figsize=(6,6))
a = fig.add_subplot(111)
a.plot(sol2[:, 0], sol2[:, 1])
a.plot(sol2[:, 2], sol2[:, 3])
# in fact, both the Earth and the Moon are moving around their center of mass,
# with the only difference being the dimensions of each trajectory.
# you can go ahead and test this, but the trajectory of the Earth is roughly
# the size of the Earth itself (in fact smaller, since the center of mass is
# inside the Earth).
```

```python
plt.close('all')

fig = plt.figure(figsize=(4,4))
a = fig.add_subplot(111)
dist = 386e6
a.set_xlim(-dist*1.1, dist*1.1)
a.set_ylim(-dist*1.1, dist*1.1)

points7 = a.scatter(sol2[0,  :2],
                    sol2[0, 2:4])

def init7():
    return points7,

def animate7(i):
    points7.set_offsets([sol2[i,  :2],
                         sol2[i, 2:4]])
    return points7,

anim = animation.FuncAnimation(fig,
                               animate7,
                               init_func = init7,
                               frames = sol2.shape[0],
                               interval = 20,
                               repeat_delay = 1000,
                               blit = True)
```

B. "Molecular dynamics"
-----------------------

(please note that this is not the way I would perform actual studies of molecular dynamics, but it's a reasonable example).

In the following, you are provided with the Hamiltonian and the initial condition for a system of "molecules". Please construct an animation of the particles interacting.

```python
nparticles = 8

position = []
momentum = []
for i in range(nparticles):
    position += [sp.Symbol('x{0}'.format(i)),
                 sp.Symbol('y{0}'.format(i))]
    momentum += [sp.Symbol('p{0}'.format(i)),
                 sp.Symbol('q{0}'.format(i))]

def VanDerWaals(r1, r2):
    dist = sp.sqrt((r1[0] - r2[0])**2 + (r1[1] - r2[1])**2)
    return dist**(-12) - dist**(-6)

def Box(r, exponent = 4):
    return ((r[0]/(nparticles))**(2*exponent) +
            (r[1]/(nparticles))**(2*exponent))

# first, put in kinetic energy term
H = sum(p**2/2 for p in momentum)

# second, put in walls
H += sum(Box(position[2*i:2*(i+1)]) for i in range(nparticles))

# third, put in pairwise interactions
for i in range(nparticles):
    for j in range(i+1, nparticles):
        H += VanDerWaals(position[2*i:2*(i+1)],
                         position[2*j:2*(j+1)])
        
bla = hamiltonian_system(position, momentum, H)
```

```python
initial_condition = np.zeros(nparticles*4, dtype = np.float)
initial_condition[0:2*nparticles:2] = \
    np.linspace(-nparticles/2, nparticles/2, nparticles)
alpha = np.random.random()*np.pi*2
initial_condition[2*nparticles  ] = 2*(np.cos(alpha)+2)
initial_condition[2*nparticles+1] = 2*np.sin(alpha)
```

```python
plt.close('all')
dt, err = get_evdt(initial_condition,
                   2.**(-4), 8, 8,
                   cRK,
                   bla.numpy_rhs)
fig = plt.figure()
a = fig.add_subplot(111)
a.plot(dt, np.average(err, axis = 1), marker = '.')
a.set_xscale('log')
a.set_yscale('log')
```

```python
sol3 = cRK(bla.numpy_rhs,
           dt = 2.**(-8),
           N = 2**12,
           x0 = initial_condition)


fig = plt.figure(figsize=(6,6))
a = fig.add_subplot(111)
a.plot(sol3[:, 0], sol3[:, 1])
a.plot(sol3[:, 2], sol3[:, 3])
```

```python
plt.close('all')

fig = plt.figure(figsize=(4,4))
a = fig.add_subplot(111)
dist = 1.25*nparticles
a.set_xlim(-dist*1.1, dist*1.1)
a.set_ylim(-dist*1.1, dist*1.1)

points8 = a.scatter(sol3[0, 0:2*nparticles:2],
                    sol3[0, 1:2*nparticles:2])

def init8():
    return points8,

def animate8(i):
    points8.set_offsets(sol3[i*8, :2*nparticles].T)
    return points8,

anim = animation.FuncAnimation(fig,
                               animate8,
                               init_func = init8,
                               frames = sol3.shape[0]/8,
                               interval = 40,
                               repeat_delay = 1000,
                               blit = True)
```
